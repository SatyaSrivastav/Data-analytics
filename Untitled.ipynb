```python
import pandas as pd
df = pd.read_excel(r"C:\Users\sriva\Downloads\Customer Call List.xlsx")


df["Last_Name"]=df["Last_Name"].str.replace("NaN","")
df["Last_Name"]=df["Last_Name"].str.replace("...","")
df["Last_Name"]=df["Last_Name"].str.replace("_","")
df=df.drop_duplicates()
df=df.fillna("")

import re
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: re.sub(r'\D', '', str(x)))
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('No','N')
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('Yes','Y')
df["Paying Customer"]=df["Paying Customer"].str.replace('N/a','')
df['Street_Address'].str.replace('N/a','')

df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',expand=True)

df=df.drop('Address',axis=1)

for x in df.index:
    if df.loc[x,'Phone_Number']=='':
        df.drop(x,inplace=True)
        
df.reset_index(drop=True)
```
